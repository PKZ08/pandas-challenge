```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to load
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing file and store into Pandas data frame
purchaseData_df = pd.read_csv(file_to_load)

purchaseData_df.head()
```

```python
#just to understand the df 
purchaseData_df.shape
```

```python
#just to understand the df 
purchaseData_df.count()
```

# Player Count
## Display the total number of players

```python
#Display the total nb of players

totalPlayers = purchaseData_df['SN'].nunique() #for Counting non-NA/null observations in the Series

# Constructing DataFrame from a dictionary.

pd.DataFrame({
    "Total Players": [totalPlayers]
})    
```

# Purchasing Analysis (Total)

### - Run basic calculations to obtain number of unique items, average price, etc
### - Create a summary data frame to hold the results
### - Optional: give the displayed data cleaner formatting
### - Display the summary data frame


```python
# Run basic calculations to obtain number of unique items, average price, etc

nbUniqueItems = purchaseData_df['Item ID'].nunique()
avgPrice = purchaseData_df["Price"].mean()
totalRevenue = purchaseData_df['Price'].sum()
nbPurchases = purchaseData_df['Purchase ID'].count()
```

```python
# Create a summary data frame to hold the results

summaryData_df = pd.DataFrame({
    "Number of Unique Items": [nbUniqueItems],
    "Average Price": [avgPrice],
    "Number of Purchases": [nbPurchases],
    "Total Revenue": [totalRevenue]
})

summaryData_df.head()
```

```python
#Optional: give the displayed data cleaner formatting

summaryData_df["Average Price"] =summaryData_df["Average Price"].map("${:,.2f}".format)
summaryData_df["Total Revenue"] =summaryData_df["Total Revenue"].map("${:,.2f}".format)
```

```python
#Display the summary data frame

summaryData_df.head()
```

# Gender Demographics

### - Percentage and count of Male Players
### - Percentage and count of Female Players
### - Percentage and count of Other / Non-Disclosed

```python
#Percentage and count of Male Players

groupbyGender_df = purchaseData_df.groupby(["Gender"])

#Creating the summary data frame

genderCount = groupbyGender_df["SN"].nunique()


genderPercentage = 100 * (genderCount / totalPlayers)

genderDemographics_df = pd.DataFrame({
    "Total Count": genderCount,
   "Percentage of players":genderPercentage
                                     })

genderDemographics_df.head()
```

```python
#Giving format

genderDemographics_df["Percentage of players"] = genderDemographics_df["Percentage of players"].map("{:.1f}%".format)

genderDemographics_df[["Total Count", "Percentage of players"]].sort_values(["Total Count"], ascending = False)

```

# Purchasing Analysis (Gender)

```python
#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender

purchaseCountGender = groupbyGender_df["Purchase ID"].count()
avgPurchasePriceGender = groupbyGender_df["Price"].mean()
totalPurchaseValueGender = groupbyGender_df["Price"].sum()
avgTotalPurchasePersonGender = totalPurchaseValueGender / genderCount


```

# Age Demographics 

### - Establish bins for ages
### - Categorize the existing players using the age bins. Hint: use pd.cut()
### - Calculate the numbers and percentages by age group
### - Create a summary data frame to hold the results
### - Optional: round the percentage column to two decimal points
### - Display Age Demographics Table

```python
# Establish bins for ages
binsAges = [0, 9, 14, 19, 24, 29, 34, 39, 100]
ageLabel = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

```

```python
# Categorize the existing players using the age bins. Hint: use pd.cut()
# Documentation for pd.cut() https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.cut.html


purchaseData_df["Age Labels"] = pd.cut(purchaseData_df["Age"], binsAges, labels=ageLabel)


groupedAge_df = purchaseData_df.groupby(["Age Labels"])

groupedAge_df.head()
```

```python
# Calculate the numbers and percentages by age group
ageCount = groupedAge_df["SN"].nunique()
agePercentage = 100* (ageCount / totalPlayers) 
```

```python
# Create a summary data frame to hold the results
ageDeomographics_df = pd.DataFrame({"Total Count": ageCount,
                            "Percentage of Players": agePercentage})

ageDeomographics_df.head()
```

```python
# Optional: round the percentage column to two decimal points
ageDeomographics_df["Percentage of Players"] = ageDeomographics_df["Percentage of Players"].map("{:.2f}%".format)

```

```python
# Display Age Demographics Table
ageDeomographics_df[["Total Count", "Percentage of Players"]]
```

# Most Popular Items

### - Retrieve the Item ID, Item Name, and Item Price columns
### - Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
### - Create a summary data frame to hold the results
### - Sort the purchase count column in descending order
### - Optional: give the displayed data cleaner formatting
### - Display a preview of the summary data frame

```python
# Retrieve the Item ID, Item Name, and Item Price columns
items_df = purchaseData_df[["Item ID", "Item Name", "Price"]]

items_df.head()
```

```python
# Group by Item ID and Item Name. 
groupedItems_df = items_df.groupby(["Item ID", "Item Name"])
```

```python
# Perform calculations to obtain purchase count, item price, and total purchase value
purchaseCountItem = groupedItems_df["Item ID"].count()
itemPriceItem = groupedItems_df["Price"].mean()
totalPurchaseValueItem = groupedItems_df["Price"].sum()

```

```python
# Create a summary data frame to hold the results
items_df = pd.DataFrame({"Purchase Count": purchaseCountItem,
                         "Item Price": itemPriceItem,
                         "Total Purchase Value": totalPurchaseValueItem})

items_df.head()
```

```python
# Sort the purchase count column in descending order
items_df = items_df.sort_values(["Purchase Count"], ascending=False)

items_df.head()
```

```python
# Optional: give the displayed data cleaner formatting
items_df["Item Price"] = items_df["Item Price"].map("${:,.2f}".format)
items_df["Total Purchase Value"] = items_df["Total Purchase Value"].map("${:,.2f}".format)

```

```python
# Display a preview of the summary data frame
items_df[["Purchase Count", "Item Price", "Total Purchase Value"]].head()
```

# Most Profitable Items

### - Sort the above table by total purchase value in descending order
### - Optional: give the displayed data cleaner formatting
### - Display a preview of the data frame


```python
# Sort the above table by total purchase value in descending order
# Converting the "Total Purchase Value" column back to floats
items_df["Total Purchase Value"] = items_df["Total Purchase Value"].apply(lambda x: x.replace('$', '').replace(',', '')).astype('float')



items_df = items_df.sort_values(["Total Purchase Value"], ascending=False)


```

```python
items_df.head()
```

```python

# Optional: give the displayed data cleaner formatting
items_df["Total Purchase Value"] = items_df["Total Purchase Value"].map("${:,.2f}".format)

# Display a preview of the data frame
items_df[["Purchase Count", "Item Price", "Total Purchase Value"]].head()
```
